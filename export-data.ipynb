#### to_csv 

```python
import pandas as pd
import zipfile

# Path to the ZIP file
zip_file_path = 'deliveries.csv.zip'

with zipfile.ZipFile(zip_file_path, 'r') as z:
    print(z.namelist())

# Open the ZIP file and extract the desired CSV file
with zipfile.ZipFile(zip_file_path, 'r') as z:
    with z.open('deliveries.csv') as f:
        df = pd.read_csv(f)
# Display the first few rows of the dataframe
df.head()

```

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
temp_df
```

```python
temp_df.to_csv('batsman_run.csv',index=False)
```

```python
df.pivot_table(index='batsman', columns='bowling_team', values='batsman_runs', aggfunc='sum').to_csv('batsman_vs_team.csv')
```
